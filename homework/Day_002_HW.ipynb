```python
random_fetch = app_train['ENTRANCES_MODE'].sample(n=3, random_state=1)
print('%s' % (random_fetch))
```

```python
random_fetch = app_train.sample(frac=0.5, replace=True, random_state=2)
print('%s' % (random_fetch))
```

```python
random_fetch = app_train.sample(n=3, replace=1, random_state =0, axis =1)
print('%s' % (random_fetch))
```
